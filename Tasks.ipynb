### Task 2

1) Using visitor-interests.csv dataset, from the first 50 row extract Browser, OS, Device data from 'UserAgent' column and use it to create new columns for Browser, OS, Device.

```python
df = pd.read_csv('visitor-interests.csv')
```

```python
df = df.loc[:50]
df
```

```python
from ip2geotools.databases.noncommercial import DbIpCity
df['Browser'] = df['UserAgent'].apply(lambda uad: ua.parse(uad).browser.family)
df['OS'] = df['UserAgent'].apply(lambda uad: ua.parse(uad).os.family)
df['Device'] = df['UserAgent'].apply(lambda uad: ua.parse(uad).device.family)
```

```python
df
```

2) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column.

```python
df['latitude'] = df['IP'].apply(lambda ip: DbIpCity.get(ip, api_key='free').latitude)
df['longitude'] = df['IP'].apply(lambda ip: DbIpCity.get(ip, api_key='free').longitude)
```

```python
df
```

### Task 3

1) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column and use it to create new 'Dist_From_Egypt' column calculating distance with Egypt Lat/Lonng, you can get this location from google.

```python
from geopy.distance import geodesic
```

```python
Egypt_location = (26.820553 ,30.802498)
```

```python
df.dropna(axis = 0, inplace=True)
```

```python
df
```

```python
df['Dist_From_Egypt']=df[['latitude','longitude']].apply(lambda x: geodesic((x[0],x[1]), Egypt_location).kilometers, axis=1)
```
